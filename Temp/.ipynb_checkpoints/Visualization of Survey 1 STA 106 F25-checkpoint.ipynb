## Major

The visualization doesn't look so great... Why?

```r
options(repr.plot.width=9, repr.plot.height=18)
survey_data %>%
  group_by(Major) %>%
  summarize(Count = n()) %>%
  ggplot(aes(x = reorder(Major, Count), y = Count)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs(title = "Distribution of Majors", x = "Major", y = "Count")
```

Merging variants of majors

```r
unique_majors <- unique(survey_data$Major)
print(unique_majors[1:10])
```

```r
survey_data <- survey_data %>%
  mutate(Major = tolower(Major)) %>%  # Convert to lowercase for consistency
  mutate(Major_cleaned = case_when(
    grepl("stat", Major) ~ "Statistics",
    grepl("quantitative psychology", Major) ~ "Psychology",
    grepl("psychology", Major) ~ "Psychology",
    grepl("computer", Major) & grepl("stat", Major) ~ "Computer Science and Statistics",
    grepl("computer", Major) ~ "Computer Science",
    grepl("econ", Major) ~ "Economics",
    grepl("systems", Major) | grepl("synthetic biology", Major) ~ "Biological Sciences",
    grepl("cognitive", Major) & grepl("stat", Major) ~ "Cognitive Science and Applied Statistics",
    TRUE ~ "Other"
  ))
```

```r
options(repr.plot.width=9, repr.plot.height=9)
survey_data %>%
  group_by(Major_cleaned) %>%
  summarize(Count = n()) %>%
  ggplot(aes(x = reorder(Major_cleaned, Count), y = Count)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs(title = "Distribution of Majors, STA 106 F24", x = "Major", y = "Count")+
    theme(axis.text=element_text(size=15))
```

## STA 108 .. or not

```r
options(repr.plot.width=9, repr.plot.height=9)
survey_summary <- table(survey_data$STA108)

# Now, we can draw the pie chart
pie(survey_summary,
    main = "STA108",
    col = c("lightblue", "lightgreen", "salmon"), # Colors for each slice
    labels = paste(names(survey_summary), "\n", survey_summary) # Adding labels with counts
)
```

```r
options(repr.plot.width=9, repr.plot.height=9)
survey_data$Major_cleaned <- factor(survey_data$Major_cleaned)
ggplot(survey_data, aes(x = reorder(Major_cleaned, Major_cleaned, FUN = length), fill = STA108)) +
  geom_bar(position = "stack") +
  labs(title = "Major and STA 108, Fall 25",
       x = "Major",
       y = "Count",
       fill = "STA108") +
    coord_flip() +
  theme_minimal()+
    theme(axis.text=element_text(size=15))
ggsave("./stackedbar_F25.jpg", device = "jpg")
```

## Reason

```r
draw.word.cloud<-function(dat){
    goals_corpus <- Corpus(VectorSource(dat))
goals_corpus <- tm_map(goals_corpus, content_transformer(tolower))
goals_corpus <- tm_map(goals_corpus, removePunctuation)
goals_corpus <- tm_map(goals_corpus, removeNumbers)
goals_corpus <- tm_map(goals_corpus, removeWords, stopwords("english"))
goals_corpus <- tm_map(goals_corpus, stripWhitespace)
    dtm <- TermDocumentMatrix(goals_corpus)
m <- as.matrix(dtm)
v <- sort(rowSums(m), decreasing = TRUE)
d <- data.frame(word = names(v), freq = v)
set.seed(1234)
#wordcloud(words = d$word, freq = d$freq, min.freq = 1, random.order = FALSE)
wordcloud(words = d$word, freq = d$freq, min.freq = 1, random.order = FALSE, scale = c(4, 0.5))

}
```

```r
options(repr.plot.width=9, repr.plot.height=9)

draw.word.cloud(survey_data$Reason)
```

```r
options(repr.plot.width=9, repr.plot.height=9)

draw.word.cloud(survey_data$Goal)
```
